```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
game = pd.read_csv('../data/clean_game_data.csv')
game
#DATAFRAME CLEANSING (STANDARDIZED NAMES) NEEDED
```

```python
game["point_diff"] = game["home_points"] - game["away_points"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X = game[["home.totalYards","Home_possession_time","Home_Offensive_TDs","home.turnovers"]].dropna()
y = game.loc[X.index]['point_diff']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(model.intercept_)
print(model.coef_)
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```
