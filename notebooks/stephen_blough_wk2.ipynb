```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.insert(0, "../")
from eff_calc import efficiencyFormatter
```

```python
games = pd.read_csv('../data/games.csv')
games
```

```python
games_data = pd.read_csv("../data/game_stats.csv")
games_data
```

```python
data = games.merge(games_data, right_on = 'id', left_on = 'id')
data
```

```python
data['Home_Offensive_TDs'] = data['home.rushingTDs'] + data['home.passingTDs']
data['Away_Offensive_TDs'] = data['away.rushingTDs'] + data['away.passingTDs']
```

```python
data = efficiencyFormatter(data)
```

```python
new_data = data[['season', 'week', 'home_team', 'away_team', 'home.totalYards', 'away.totalYards', 'away.possessionTime', 'home.possessionTime', 'Home_Offensive_TDs', 'Away_Offensive_TDs', 'home.turnovers', 'away.turnovers', 'home.thirdDownEff', 'home.fourthDownEff', 'away.thirdDownEff', 'away.fourthDownEff', 'away_points', 'home_points']].dropna()
```

```python
for i in new_data['home.possessionTime']:
    m, s = i.split(":")
    new_data['Home_possession_time'] = int(m) * 60 + int(s)
for j in new_data['away.possessionTime']:
    m, s = j.split(":")
    new_data['Away_possession_time'] = int(m) * 60 + int(s)
```

```python
clean_data = new_data[['season', 'week', 'home_team', 'away_team', 'home.totalYards', 'away.totalYards', 'Home_possession_time', 'Away_possession_time','Home_Offensive_TDs', 'Away_Offensive_TDs', 'home.turnovers', 'away.turnovers', 'home.thirdDownEff', 'home.fourthDownEff', 'away.thirdDownEff', 'away.fourthDownEff', 'home_points', 'away_points']].dropna()
```
